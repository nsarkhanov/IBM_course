```python
df_can = pd.read_excel("Canada.xlsx",sheet_name='Canada by Citizenship',
                       skiprows=range(20),
                       skipfooter=2)
```

```python
df_can=df_can.dropna(axis=1)
df_can.drop(['AREA','REG','DEV','Type','Coverage'], axis=1, inplace=True)
```

```python
df_can
```

```python
colum_list=df_can.columns[4:]
df_can['Total']=df_can[colum_list].sum(axis=1)
df_can
```

```python
df_can.sort_values(['Total'],ascending=False,axis=0,inplace=True)
df_can
```

```python
df=df_can.set_index('OdName')

df
```

```python
years=list(map(int,range(1980,2014)))
years
```

```python
df[years]
```

```python

df_10=df.head(10)
df_10
```

```python
df_=df_10[years].transpose()
df_.plot(kind="area",
         stacked=False,
         figsize=(20,10))
plt.title("Top 10  migration country in Canada ")
plt.xlabel("Years")
plt.ylabel("Number of migrations")
plt.show()
```
